```python
df_mouse_expression = pd.read_csv(mouse_expression, sep='\t')
```

```python
df_mouse_expression
```

```python
gene_given = list(df_mouse_expression["Gene.symbol"])
gene_human = list(df_expression.columns)


# len(gene_given)

metformin_induced_genes = []
processed_count = 0
count = 0
for i in gene_given:
    count += 1
    if (not isinstance(i, str)):
        # print(f"{i} --> {type(i)}")
        continue 
    if str.upper(i) in gene_human:
        processed_count += 1
        metformin_induced_genes.append(str.upper(i))

print(f"Total: {count} || Processed: {processed_count}")
print(f"Genes related to Metformin (count: {len(metformin_induced_genes)})\n\t{metformin_induced_genes}")
```
