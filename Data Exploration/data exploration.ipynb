## Imports and Ingestion

```python
import os
import numpy as np
import pandas as pd
import sys
import matplotlib.pyplot as plt
```

```python
human_expression_path = f"{os.getcwd()}/../BE700_human_expression.(1) (1) (1).csv"
human_pheno_path = f"{os.getcwd()}/../BE700_human_pheno (1) (2) (1) (2).csv"
```

## Preprocessing
Human Phenotype data consists of both continuous data and class based data, among which there are non-differentiable data (only one class, e.g. all samples are `Homo sapiens` for `organism_ch1` variable). We will first remove those columns.

```python
df_human_expression = pd.read_csv(human_expression_path, index_col=0).T
df_human_pheno = pd.read_csv(human_pheno_path, index_col=0)
```

```python
print(f"Columns in Human Phenotype: {list(df_human_pheno.columns)}")
print()
single_value_columns = df_human_pheno.apply(lambda col: col.nunique() == 1)
columns_with_one_value = single_value_columns[single_value_columns].index.tolist()
print(f"Columns with only 1 value in Human Phenotype: {columns_with_one_value}")
```

```python
## Drop columns above
df_human_pheno = df_human_pheno.drop(columns=columns_with_one_value)
```

```python
## Drop 'characteristics' columns and other replicates
columns_characteristics = []
for i in list(df_human_pheno.columns):
    if ("characteristics" in i):
        columns_characteristics.append(i)
df_human_pheno = df_human_pheno.drop(columns=columns_characteristics)
df_human_pheno = df_human_pheno.drop(columns=["title"])
```

```python
## Drop non-biologically relevant data columns
df_human_pheno = df_human_pheno.drop(columns=["geo_accession", "last_update_date", "description", "supplementary_file"])
```

```python
## Process description into boolean column
df_human_pheno = df_human_pheno.assign(diabete_2=df_human_pheno.apply(lambda x: 0 if "without diabetes" in x["description.1"] else 1, axis=1))
# df_human_pheno.apply(lambda x: print("negative") if "without diabetes" in x["description.1"] else print("positive"), axis=1)
df_human_pheno = df_human_pheno.drop(columns=["description.1"])
```

```python
df_human_pheno.replace("missing value", np.nan, inplace=True)
```

```python
df_human_pheno["waist hip ratio:ch1"] = df_human_pheno["waist hip ratio:ch1"].astype(float)
```

## Data Exploration (Human Phenotype)

#### NA Value Counting

```python
na_count = {
    "valid": np.array(df_human_pheno.count()),
    "na": np.array(df_human_pheno.isna().sum())
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(len(list(df_human_pheno)))

for type_, count in na_count.items():
    val = ax.bar(
        df_human_pheno.columns,  
        count, 
        width, 
        label=type_, 
        bottom=bottom
    )
    ax.bar_label(val, label_type='center')
    bottom += count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")
ax.set_ylabel("Count")
ax.set_ylim([0, 75])
# ax.set_xlabel("Columns")
ax.legend(loc="upper right")
plt.xticks(rotation=45, ha="right")  
plt.tight_layout()

plt.show()
```

#### Distribution Analysis

```python
fig, ax = plt.subplots(figsize=(16,12), nrows=int(len(df_human_pheno.columns)/4), ncols=int(len(df_human_pheno.columns)/(len(df_human_pheno.columns)/4)))

index = 0
for i in range(int(len(df_human_pheno.columns)/4)):
    for j in range(int(len(df_human_pheno.columns)/(len(df_human_pheno.columns)/4))):        
        ax[i, j].hist(df_human_pheno[list(df_human_pheno.columns)[index]], color='lightgreen', ec='black', bins=50)
        ax[i, j].set_title(list(df_human_pheno.columns)[index])
        ax[i, j].set_ylabel("Count")
        ax[i, j].tick_params(labelrotation=45)
        index += 1

fig.tight_layout()
```
